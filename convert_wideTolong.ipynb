```python
import pandas as pd
import dask.dataframe as dd
```

```python
df_original = pd.read_csv('original.csv')
```

```python
df_original
```

```python
df_original = df_original.drop(columns=['RESID_STATE_90',
       'AGE_90', 'EMPLOYMENT_90', 'EDUC_LEVEL_90', 'RESID_STATE_91', 'AGE_91',
       'EMPLOYMENT_91', 'EDUC_LEVEL_91', 'RESID_STATE_92', 'AGE_92',
       'EMPLOYMENT_92', 'EDUC_LEVEL_92', 'RESID_STATE_93', 'AGE_93',
       'EMPLOYMENT_93', 'EDUC_LEVEL_93', 'RESID_STATE_94', 'AGE_94',
       'EMPLOYMENT_94', 'EDUC_LEVEL_94', 'RESID_STATE_95', 'AGE_95',
       'EMPLOYMENT_95', 'EDUC_LEVEL_95', 'RESID_STATE_96', 'AGE_96',
       'EMPLOYMENT_96', 'EDUC_LEVEL_96', 'RESID_STATE_97', 'AGE_97',
       'EMPLOYMENT_97', 'EDUC_LEVEL_97', 'RESID_STATE_99', 'AGE_99',
       'EMPLOYMENT_99', 'EDUC_LEVEL_99'])
```

```python
df_original.columns
```

```python
long_df = pd.DataFrame(columns=['# Child_BIRTHS_ALL_YEAR', 'MARITAL_STATUS_ALL_YEAR', 
                                'RESID_STATE', 'AGE', 'EMPLOYMENT', 'EDUC_LEVEL', 'YEAR'])

# Loop through each set of 10 wide columns and concatenate the values
for i in range(1, 20, 2):
    temp_df = df_original[['# Child_BIRTHS_ALL_YEAR', 'MARITAL_STATUS_ALL_YEAR', 
                  f'RESID_STATE_{i:02}', f'AGE_{i:02}', f'EMPLOYMENT_{i:02}', f'EDUC_LEVEL_{i:02}']].copy()
    temp_df['YEAR'] = i
    temp_df.columns = ['# Child_BIRTHS_ALL_YEAR', 'MARITAL_STATUS_ALL_YEAR', 
                       'RESID_STATE', 'AGE', 'EMPLOYMENT', 'EDUC_LEVEL', 'YEAR']
    
    # Ensure the correct data types
    temp_df['RESID_STATE'] = temp_df['RESID_STATE'].astype(int)
    temp_df['AGE'] = temp_df['AGE'].astype(int)
    temp_df['EMPLOYMENT'] = temp_df['EMPLOYMENT'].astype(int)
    temp_df['EDUC_LEVEL'] = temp_df['EDUC_LEVEL'].astype(int)
    
    long_df = pd.concat([long_df, temp_df], ignore_index=True)

# Reset index
long_df.reset_index(drop=True, inplace=True)
```

```python
long_df.to_csv('long_original.csv', index=False)
```
